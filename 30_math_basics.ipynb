```python
import numpy as np
```

```python
from pylab import plt
plt.style.use('seaborn-v0_8')
%config InlineBackend.figure_format = 'svg'
```

```python
def R2(yn, yn_):
    c = np.corrcoef(yn, yn_)[0, 1]
    return c ** 2
```

## Importing Data 

```python
import urllib
```

### The File

```python
url = 'https://hilpisch.com/aiif_eikon_eod_data.csv'
```

```python
fn = '../../../data/'+url.split('/')[-1]
```

```python
urllib.request.urlretrieve(url, fn)
```

### Reading from Disk

```python
raw = np.genfromtxt(fn, delimiter=',', names=True)
```

```python
raw[:5]
```

```python
raw[-5:][['AAPLO', 'MSFTO']]
```

## Regression Analysis 1

### Extracting a Time Series

```python
spx = raw['SPX']
```

```python
spx
```

```python
np.isnan(spx)
```

```python
~np.isnan(spx)
```

```python
spx = spx[~np.isnan(spx)]
```

```python
spx[:15]
```

```python
plt.plot(spx);
```

### Simple Linear Regression

```python
len(spx)
```

```python
x = np.arange(len(spx))
```

```python
reg = np.polyfit(x, spx, deg=1)
```

```python
reg
```

```python
plt.plot(x, spx)
plt.plot(x, np.polyval(reg, x), 'r--');
```

```python
R2(spx, np.polyval(reg, x))
```

## Regression Analysis 2

### Extracting a Time Series

```python
vix = raw['VIX']
```

```python
vix = vix[~np.isnan(vix)]
```

```python
vix[:15]
```

```python
vix.mean()
```

```python
vix.min()
```

```python
vix.max()
```

```python
plt.plot(vix);
```

### Simple Linear Regression

```python
len(vix)
```

```python
x = np.arange(len(vix))
```

```python
reg = np.polyfit(x, vix, deg=1)
```

```python
reg
```

```python
plt.plot(x, vix)
plt.plot(x, np.polyval(reg, x), 'r--');
```

```python
R2(vix, np.polyval(reg, x))
```

## Regression Analysis 3

## Two Time Series

```python
np.set_printoptions(suppress=True)
```

```python
r_spx = np.log(spx[1:] / spx[:-1])
```

```python
r_spx
```

```python
r_vix = np.log(vix[1:] / vix[:-1])
```

```python
r_vix
```

```python
plt.scatter(x=r_spx, y=r_vix, marker='.', c='b', alpha=0.5);
```

### Simple Linear Regression

```python
reg = np.polyfit(r_spx, r_vix, deg=1)
```

```python
reg
```

```python
plt.scatter(r_spx, r_vix, marker='.', c='b', alpha=0.5)
plt.plot(r_spx, np.polyval(reg, r_spx), 'r--');
```

```python
R2(r_vix, np.polyval(reg, r_spx))
```

```python
np.corrcoef(r_spx, r_vix)
```
