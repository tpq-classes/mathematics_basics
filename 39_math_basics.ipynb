### `random` vs. `numpy.random`

```python
import sys
import random
from numpy.random import default_rng
```

```python
random.gauss(0, 1)
```

```python
%timeit random.gauss(0, 1)
```

```python
rng = default_rng(100)
```

```python
rng.standard_normal()
```

```python
%timeit rng.standard_normal()
```

```python
I = 5000
```

```python
%time mat = [[random.gauss(0, 1) for j in range(I)] \
             for i in range(I)]  
```

```python
mat[0][:5]
```

```python
%time sum([sum(l) for l in mat])  
```

```python
sum([sys.getsizeof(l) for l in mat])  
```

```python
%time mat = np.random.standard_normal((I, I))  # legacy way
```

```python
sys.getsizeof(mat)  
```

## Loops

### Python

```python
import random
```

```python
def average_py(n):
    s = 0  
    for i in range(n):
        s += random.random()  
    return s / n
```

```python
n = 10000000  
```

```python
%time average_py(n)  
```

```python
%timeit average_py(n) 
```

```python
%timeit sum([random.random() for _ in range(n)]) / n  
```

### NumPy 

```python
import numpy as np
```

```python
def average_np(n):
    s = np.random.random(n)  
    return s.mean()  
```

```python
%time average_np(n)
```

```python
%timeit average_np(n)
```

```python
def average_np_new(n):
    s = rng.random(n)  
    return s.mean()  
```

```python
%time average_np_new(n)
```

```python
%timeit average_np_new(n)
```

```python
s = np.random.random(n)
s.nbytes  
```

### Cython

```python
%load_ext Cython
```

```python
%%cython -a
import random  
def average_cy1(int n):  
    cdef int i  
    cdef float s = 0  
    for i in range(n):
        s += random.random()
    return s / n
```

```python
%time average_cy1(n)
```

```python
%timeit average_cy1(n)
```

```python
%%cython -a
from libc.stdlib cimport rand  
cdef extern from 'limits.h':  
    int INT_MAX  
cdef int i
cdef float rn
for i in range(5):
    rn = rand() / INT_MAX  
    print(rn)
```

```python
%%cython -a
from libc.stdlib cimport rand  
cdef extern from 'limits.h':  
    int INT_MAX  
def average_cy2(int n):
    cdef int i
    cdef float s = 0
    for i in range(n):
        s += rand() / INT_MAX  
    return s / n
```

```python
%time average_cy2(n)
```

```python
%timeit average_cy2(n)
```
