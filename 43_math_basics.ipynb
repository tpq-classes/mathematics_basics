### `Series` Object

```python
import pandas as pd
```

```python
s = pd.Series(l)
```

```python
s = pd.Series(v)
```

```python
s
```

```python
type(s)
```

```python
range(0, 10, 1)
```

```python
list(range(0, 10, 1))
```

```python
s.index
```

```python
s.dtype  # ... has a unique dtype
```

```python
s.astype(float)
```

```python
s.tolist()
```

```python
t = pd.Series(range(5, 15))
```

```python
t
```

```python
s + t  # vector addition
```

```python
a = 2  # scalar
```

```python
a * s  # scalar multiplication
```

```python
a * t  # scalar multiplication
```

```python
s.to_string()
```

```python
print(t.to_string())
```

## Creating `Series` Objects

```python
pd.Series((0, 1, 2, 3))  # from tuple
```

```python
pd.Series([0, 1, 2, 3])  # from list
```

```python
pd.Series(range(5))  # from range object
```

```python
pd.Series(np.arange(0, 5))  # from ndarray object
```

```python
pd.Series(np.linspace(0, 1, 5))  # from ndarray object
```
