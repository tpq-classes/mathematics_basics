## GroupBy Operations

```python
df['Quarter'] = ['Q1', 'Q1', 'Q1',
                 'Q2', 'Q2', 'Q2',
                 'Q3', 'Q3', 'Q3',
                 'Q4', 'Q4', 'Q4']
df
```

```python
groups = df.groupby('Quarter')  
```

```python
groups.size()
```

```python
groups.mean()
```

```python
groups.max()
```

```python
groups.aggregate([min, max]).round(2)  
```

```python
df['Odd_Even'] = ['Odd', 'Even', 'Odd', 'Even',
                  'Odd', 'Even', 'Odd', 'Even',
                  'Odd', 'Even', 'Odd', 'Even',]
```

```python
df
```

```python
groups = df.groupby(['Quarter', 'Odd_Even'])
```

```python
groups.size()
```

```python
groups[['No1', 'No3']].aggregate([sum, np.mean])
```

## Complex Selection

```python
data = rng.standard_normal((10, 2))  # new way
```

```python
df = pd.DataFrame(data, columns=['x', 'y'])  
```

```python
df.describe()
```

```python
df.info()  
```

```python
df.head()  
```

```python
df.tail()  
```

```python
df['x'] > 0.5  
```

```python
(df['x'] > 0) & (df['y'] < 0)  
```

```python
(df['x'] > 0) | (df['y'] < 0)  
```

```python
df['x'] > 0
```

```python
df[df['x'] > 0]
```

```python
df.query('x > 0')
```

## Concatenation, Joining and Merging

```python
df1 = pd.DataFrame(['100', '200', '300', '400'], 
                    index=['a', 'b', 'c', 'd'],
                    columns=['A',])
```

```python
df1
```

```python
df2 = pd.DataFrame(['200', '150', '50'], 
                    index=['f', 'b', 'd'],
                    columns=['B',])
```

```python
df2
```

#### Concatenation

```python
pd.concat((df1, df2), sort=False) 
```

```python
pd.concat((df1, df2), ignore_index=True, sort=False)  
```

#### Joining

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
df1.join(df2, how='left')  # default
```

```python
df1.join(df2, how='right') 
```

```python
df1.join(df2, how='inner')  
```

```python
df1.join(df2, how='outer')  
```

```python
df = pd.DataFrame()
```

```python
df['A'] = df1['A']  
```

```python
df
```

```python
df['B'] = df2
```

```python
df
```

```python
df = pd.DataFrame({'A': df1['A'], 'B': df2['B']})  
```

```python
df
```

#### Merging

```python
c = pd.Series([250, 150, 50], index=['b', 'd', 'c'])
```

```python
c
```

```python
df1['C'] = c
```

```python
df2['C'] = c
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2) 
```

```python
pd.merge(df1, df2, on='C')  
```

```python
pd.merge(df1, df2, how='outer')  
```

```python
pd.merge(df1, df2, left_on='A', right_on='B')
```

```python
pd.merge(df1, df2, left_on='A', right_on='B', how='outer')  
```

```python
pd.merge(df1, df2, left_index=True, right_index=True)
```

```python
pd.merge(df1, df2, on='C')
```

```python
pd.merge(df1, df2, on='C', how='outer')
```

```python
pd.merge(df1, df2, left_on='A', right_on='B')
```
