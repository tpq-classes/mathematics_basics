## Speed & Memory

```python
import sys
```

```python
N = 10 ** 7
```

```python
N
```

```python
l = range(N)
```

```python
sys.getsizeof(l)
```

```python
%time l__ = list(range(N))
```

```python
sys.getsizeof(l__)
```

```python
%time l_ = [math.sqrt(x) for x in l]
```

```python
%time s = pd.Series(np.arange(N))
```

```python
%time s_ = np.sqrt(s)
```

```python
sys.getsizeof(l_)
```

```python
type(s_)
```

```python
sys.getsizeof(s_)
```

```python
s_.nbytes
```

```python
%time sum(l_)
```

```python
%time np.sum(s_)
```

```python
%timeit np.sqrt(s)
```

```python
%timeit s.apply(np.sqrt)
```

```python
%time s.apply(math.sqrt)
```
