## Speed & Memory

```python
import sys
```

```python
import math
```

```python
N = 10 ** 7
```

```python
N
```

```python
l = range(N)
```

```python
sys.getsizeof(l)
```

```python
%time l__ = list(range(N))
```

```python
sys.getsizeof(l__)
```

```python
%time l_ = [math.sqrt(x) for x in l]
```

```python
%time v = np.arange(N)
```

```python
%time v_ = np.sqrt(v)
```

```python
sys.getsizeof(l_)
```

```python
sys.getsizeof(v_)
```

```python
v_.nbytes
```

```python
%time sum(l_)
```

```python
%time np.sum(v_)
```

```python
%timeit math.sqrt(2)
```

```python
%timeit np.sqrt(2)
```

```python
%time l_ = [math.sqrt(x) for x in l[:500000]]
```

```python
%time l_ = [np.sqrt(x) for x in l[:500000]]
```
