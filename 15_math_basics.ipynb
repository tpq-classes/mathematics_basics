```python
def mean(xn):
    return sum(xn) / len(xn)
```

```python
def var(xn):
    x_ = mean(xn)
    return mean([(x - x_) ** 2 for x in xn])
```

```python
def cov(xn, yn):
    x_ = mean(xn)
    y_ = mean(yn)
    return mean([(x - x_) * (y - y_)  for x, y in zip(xn, yn)])
```

```python
def beta(x, y):
    return cov(x, y) / var(x)
```

```python
def alpha(xn, yn):
    return mean(yn) - beta(xn, yn) * mean(xn)
```

```python
def sstot(yn):
    mu = mean(yn)
    return sum([(y - mu) ** 2 for y in yn])
```

```python
def ssres(yn, yn_):
    return sum([(y - y_) ** 2 for y, y_ in zip(yn, yn_)])
```

```python
def R2(yn, yn_):
    return 1 - ssres(yn, yn_) / sstot(yn)
```

## Regression Analysis 1

### Extracting a Time Series

```python
spx = [row[7] for row in raw[2:]]
```

```python
spx[:15]
```

```python
spx = [float(row[7]) for row in raw[2:] if row[7] != '']
```

```python
spx[:15]
```

```python
plt.plot(spx);
```

### Simple Linear Regression

```python
def simlinreg(xn, yn):
    b = beta(xn, yn)
    a = alpha(xn, yn)
    return a, b
```

```python
x = range(len(spx))
```

```python
a, b = simlinreg(x, spx)
```

```python
a, b
```

```python
plt.plot(x, spx)
plt.plot(x, [a + b * x for x in x], 'r--');
```

```python
R2(spx, [a + b * x for x in x])
```
