## Random Variable

From Wikipedia (https://en.wikipedia.org/wiki/Random_variable):

> In probability and statistics, a random variable, random quantity, aleatory variable, or stochastic variable is described informally as a variable whose values depend on outcomes of a random phenomenon. In that context, a random variable is understood as a measurable function defined on a probability space that maps from the sample space to the real numbers.

These models are then used by quantitative analysts to value options on stock prices, bond prices, and on interest rates, see Markov models. Moreover, it is at the heart of the insurance industry.

## Financial Random Variables

Applying the Monte Carlo method to simulate financial quantities.

### Stock Price (1)

Assume the following parameters for a simple stock price model:

* $S_0 = 100$, initial price
* $\sigma = 10$, standard deviation ("volatility")
* $S$, future price (normally distributed)

```python
S0, sigma = 100, 10
```

```python
S = rng.normal(S0, sigma, 1000)
```

```python
S[:10]
```

```python
S.mean()
```

```python
S.var()
```

```python
S.std()
```

```python
plt.plot(S);
```

```python
plt.plot(S, 'r.', alpha=0.5);
```

```python
plt.hist(S, bins=50);
```

### Stock Price (2)

Assume the following parameters for the Black-Scholes-Merton (1973) stock price model (geometric Brownian motion):

* $S_0 = 100$, initial price
* $\sigma = 0.2$, volatility (percent in digits)
* $r = 0.01$, constant short rate
* $T = 1$, time horizon (in year fractions)
* $S$, future price (log-normally distributed)

It holds $S = S_0 e^{(r - \sigma^2 / 2) T + \sigma \sqrt{T} z}$, with $z$ standard normally distributed.

```python
S0, sigma, r, T = 100, 0.2, 0.01, 1.0
```

```python
S = S0 * np.exp((r - sigma ** 2 / 2) * T + sigma * math.sqrt(T) * rng.standard_normal(1000))
```

```python
S[:10]
```

```python
R = np.log(S / S0)  # log returns
```

```python
S.mean()  # sample mean price
```

```python
S0 * math.exp(r * T)  # forward price
```

```python
R.var()
```

```python
R.std()  # sample volatility
```

```python
plt.plot(S);
```

```python
plt.plot(S, 'r.', alpha=0.5);
```

```python
plt.hist(S, bins=50)
plt.axvline(S.mean(), c='r', label='sample mean')
plt.legend();
```

### Europen Option Payoffs

Assume the previous stock price model and assume that a European option has a strike price of $K = 100$.

The payoff of a European options is:
* $\max(S - K, 0)$ for a European call option
* $\max(K - S, 0)$ for a European put option

```python
K = 100
```

* **stock price**: random phenomenon = nature/markets --> stock price (model)
* **option payoff**: random phenomenon = stock price --> option payoff (model)

#### European Call option

```python
C = np.maximum(S - K, 0)
```

```python
C[:10]
```

```python
# Monte Carlo Estimator
C0 = math.exp(-r * T) * C.mean()  # present value
```

```python
C0
```

```python
RC = np.log(C / C0 + 0.00001)  # log returns
```

```python
C.mean()  # future value (estimator)
```

```python
RC.var()
```

```python
RC.std()
```

```python
plt.plot(C);
```

```python
plt.plot(C, 'r.', alpha=0.5);
```

```python
plt.hist(C, bins=50);
```

```python
plt.plot(S, C, 'r.')
plt.xlabel('stock price')
plt.ylabel('option payoff');
```

```python
plt.plot(R, RC, 'r.')
plt.xlabel('stock log return')
plt.ylabel('option log return');
```

#### European Put option

```python
P = np.maximum(K - S, 0)
```

```python
P0 = math.exp(-r * T) * P.mean()
```

```python
P0
```

```python
RP = np.log(P / P0 + 0.00001)  # log returns
```

```python
P.mean()
```

```python
RP.var()
```

```python
RP.std()
```

```python
plt.plot(P);
```

```python
plt.plot(P, 'r.', alpha=0.5);
```

```python
plt.hist(P, bins=50);
```

```python
plt.plot(S, P, 'r.')
plt.xlabel('stock price')
plt.ylabel('option payoff');
```

```python
plt.plot(R, RP, 'r.')
plt.xlabel('stock log returns')
plt.ylabel('option log returns');
```

#### Comparison

```python
plt.plot(S, P, 'r.', label='put', alpha=0.5)
plt.plot(S, C, 'b.', label='call', alpha=0.5)
plt.legend();
```

```python
plt.plot(R, RP, 'r.', label='put', alpha=0.5)
plt.plot(R, RC, 'b.', label='call', alpha=0.5)
plt.legend();
```
