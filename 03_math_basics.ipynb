## Standard Functions

```python
import math
```

```python
x = range(100)
x = [_ / 10 for _ in x ]
```

```python
y = [math.sqrt(_) for _ in x]
```

```python
plt.plot(x, y)
plt.title('$y = \sqrt{x}$');
```

```python
y = [math.log(_ + 0.00001) for _ in x]
```

```python
plt.plot(x, y)
plt.title('$y = \sqrt{x}$');
```

```python
y = [math.exp(_) for _ in x]
```

```python
plt.plot(x, y)
plt.title('$y = \exp(x)$');
```

```python
y1 = [math.sin(_) for _ in x]
y2 = [math.cos(_) for _ in x]
```

```python
plt.plot(x, y1, label='$y=\sin(x)$')
plt.plot(x, y2, label='$y=\cos(x)$')
plt.legend();
```

## Custom Functions

```python
def f(x):
    return x ** 3 / 5 + 10 * x ** 0.5 - 1
```

```python
y = [f(_) for _ in x]
```

```python
plt.plot(x, y)
plt.title('$y = x^3 / 5 + 10 \sqrt{x} - 1$');
```

## CASE: Discounting Cash Flows

```python
t = range(5)
```

```python
list(t)
```

```python
c = [-100, 20, 40, 50, 80]
```

```python
c
```

```python
plt.bar(t, c);
```

```python
i = 0.1
```

```python
sdf = [(1 + 0.1) ** (-t_) for t_ in t] 
```

```python
sdf
```

```python
plt.bar(t, sdf);
```

```python
list(zip(c, sdf))
```

```python
pv = [x * y for (x, y) in zip(c, sdf)]
```

```python
pv
```

```python
plt.bar(t, c, label='cash flows')
plt.bar(t, pv, label='present values')
plt.legend();
```

```python
npv = sum(pv)
```

```python
npv
```

```python
cdf = [math.exp(-i * t_) for t_ in t] 
```

```python
sdf
```

```python
plt.bar(t, sdf, label='simple')
plt.bar(t, cdf, alpha=0.8, label='continuous')
plt.legend();
```

```python
list(zip(c, cdf))
```

```python
pv = [x * y for (x, y) in zip(c, cdf)]
```

```python
pv
```

```python
plt.bar(t, c, label='cash flows')
plt.bar(t, pv, label='present values')
plt.legend();
```

```python
npv = sum(pv)
```

```python
npv
```
