## Speed & Memory

```python
import sys
```

```python
N = 10 ** 8
```

```python
N
```

```python
%time m = np.arange(N).reshape(int(N / 10), 10)
```

```python
m.nbytes
```

```python
%time df = pd.DataFrame(m)
```

```python
sys.getsizeof(df)
```

```python
%time res = np.sqrt(m)
```

```python
%timeit res = np.sqrt(m)
```

```python
%time res = np.sqrt(df)
```

```python
%timeit res = np.sqrt(df)
```

```python
type(df.values)  # from DataFrame to ndarray
```

```python
%time res = np.sqrt(df.values)
```

```python
%timeit df.apply(np.sin)
```

```python
%timeit np.sin(m)
```
